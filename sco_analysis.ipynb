```python
clusters = {}
sco = 0
nfilt = 0
cfilt = 0
with open(sco_file) as fh:
    for line in fh:
        if line.startswith("#"):
            header = line.strip().split()
            continue
        line = line.strip().split()
        print(line)
        if int(line[1]) == int(line[0]):
            sco += 1
            if int(line[0]) > 125:
                nfilt += 1
                clusters[f"cluster_{nfilt}_{line[0]}_{line[2]}"] = line[3:]
                if float(line[2]) > 0.4:                    
                    cfilt += 1
print(sco, nfilt, cfilt)
```

```python
index = 247
genome = header[index]

cluster_to_product = {}

def ncbi_id(db, db_id_query):
    from Bio import Entrez
    Entrez.email = "ad3002@gmail.com"
    handle = Entrez.esearch(db=db, retmax=100, term=db_id_query)
    query_id = Entrez.read(handle)["IdList"]
    return query_id

def ncbi_id_summary(db, db_id):
    from Bio import Entrez
    Entrez.email = "ad3002@gmail.com"
    handle = Entrez.esummary(db=db, id=db_id)
    summary = Entrez.read(handle)
    handle.close()
    return summary

for cluster, proteins in clusters.items():
    nid = ncbi_id("protein", proteins[index])
    print(nid, cluster, proteins[index])
    if not nid:
        nid = ncbi_id("protein", proteins[index-1])
    summary = ncbi_id_summary("protein", nid[0])
    product = summary[0]["Title"]
    cluster_to_product[cluster] = product
```

```python
cluster_table = "./tables/sco_clusters.tsv"

with open(cluster_table, "w") as fw:
    fw.write("#cluster\tgenomes_number\tproteins_identity_in_clusters\tproduct\n")
    for c, p in cluster_to_product.items():
        c = c.split("_")
        cluster = "_".join(c[:2])
        genomes_number = c[2]
        prot_ident = c[3]
        fw.write(f"{cluster}\t{genomes_number}\t{prot_ident}\t{p}\n")
```
